Q. Follow these steps in a single script:

Create a list containing several duplicate fruits: ['apple', 'banana', 'orange', 'apple', 'pear', 'banana'].

Convert that list into a set to automatically remove duplicates.

Convert that set into a tuple.

Create a dictionary where the keys are the unique fruits and the values are the number of characters in each fruit's name.

Print the final dictionary.

```python
fruits = ['apple', 'banana', 'orange', 'apple', 'pear', 'banana']
unique_fruits_set = set(fruits)
print(f"Set of unique fruits: {unique_fruits_set}")

unique_fruits_tuple = tuple(unique_fruits_set)
print(f"Tuple of unique fruits: {unique_fruits_tuple}")

fruit_lengths_dict = {fruit: len(fruit) for fruit in unique_fruits_tuple}
print(f"Dictionary of fruit lengths: {fruit_lengths_dict}")
```
