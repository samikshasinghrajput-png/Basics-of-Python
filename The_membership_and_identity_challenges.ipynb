Q. Write a script that does the following:

Create a list list_a = [1, 2, 3].

Create list_b = [1, 2, 3].

Create list_c = list_a.

Use the identity operators (is, is not) to compare list_a with list_b and list_a with list_c.

Use the membership operator (in) to check if the number 2 is in list_a.

Goal: Explain why list_a == list_b is True, but list_a is list_b is False.

```python
#Concept used: Variables, Identity vs. Equality, Membership Operators.

list_a = [1, 2, 3]
list_b = [1, 2, 3]
list_c = list_a

print(f"list_a is list_b: {list_a is list_b}")
print(f"list_a is not list_b: {list_a is not list_b}")
print(f"list_a is list_c: {list_a is list_c}")
print(f"list_a is not list_c: {list_a is not list_c}")
print(f"list_a == list_b: {list_a == list_b}")

print(f"2 is in list_a: {2 in list_a}")
```
